```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv")
```

```python
df
```

```python
df.describe()
```

```python
#Train-test split for splitting our dataset
from sklearn.model_selection import train_test_split
```

```python
#Define X and Y variables
X=df.loc[:,df.columns!="Spending Score (1-100)"]
y=df["Spending Score (1-100)"]
```

```python
print(X.shape)
print(y.shape)
```

```python
#Partioning the data
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#LAZYPREDICT
import lazypredict
```

REGRESSOR

```python
#import Regression class from lazypredict
from lazypredict.Supervised  import LazyRegressor
```

```python
#Define model using LazyRegressor
model_Reg=LazyRegressor(verbose=0,ignore_warnings=True,predictions=True)
```



1.   Models=model’s variable contains every model accuracy in it along with some other important information.

2.   List item



```python

#Fitting the models along with predicting the outputs of each model
models,predictions=model_Reg.fit(X_train,X_test,y_train,y_test)
```

As you can see, It has implemented 42 ML models on my regression problem

```python
models
```

Using predictions ,we can create confusion matrix

CLASSIFICATION

```python
#Import Classification class from lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
#Define model using LazyClassifier
model_Classy=LazyClassifier(verbose=0,ignore_warnings=True,predictions=True)
#Fitting the models along with predicting the outputs of each model
models1,predictions1=model_Classy.fit(X_train,X_test,y_train,y_test)
```
